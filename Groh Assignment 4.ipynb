```python
#Imports

import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("data.csv", header = 0, index_col = 0)

data.info()
```

```python
#Histogram of Fuel Prices

plt.hist(data['Fuel_Price'])
plt.xlabel("Fuel Price ($)", labelpad=14)
plt.ylabel("Frequency", labelpad=14)
plt.title("Distribution of Fuel Price", y=1.015, fontsize=22);
```

```python
#Bar chart of Unemployment

data['Unemployment'].plot(kind = 'bar')
plt.ylabel("Unemployment", labelpad=14)
plt.title("Unemployment Over Time", y=1.015, fontsize=22)
```

```python
# Line Chart

data['Temperature'].plot(kind = 'line')
plt.ylabel (' Temperature ', labelpad=14)
plt.title('Temperature Over Time', y=1.015, fontsize=22)
```

```python
#Area Chart

plt.stackplot(data['Unemployment'], data['dept_1 sales'], data['dept_2 sales'], data['dept_3 sales'])
plt.legend(['Sales 1', 'Sales 2', 'Sales 3'])
plt.ylabel ('Sales')
plt.xlabel ('Unemployment')
plt.title ('Sales Departments vs Unemployment')
```

```python
# Scatter Plot

plt.scatter(data['Temperature'], data['Fuel_Price'])
plt.xlabel ('Temperature')
plt.ylabel ('Fuel Price')
plt.title ('Fuel Price vs Temperature')
```
